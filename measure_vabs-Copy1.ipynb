```python
# Absorption lines of interest
# gf-weighted mean wavelengths (Branch et al. 1983 -- for SN Ia lines)
def select_line(line):
    if line == 'CaII':
        rest_wavelength = 3945.
        w0 = 3550
        w1 = 3900
    elif line == 'SIIblue':
        rest_wavelength = 5454.
        w0 = 5000
        w1 = 5650
    elif line == 'SIIred':
        rest_wavelength = 5640.
        w0 = 5000
        w1 = 5650
    elif line == 'SiII':
        rest_wavelength = 6355.
        w0 = 6000
        w1 = 6350
    elif line == 'HeI4471':
        rest_wavelength = 4471.
        w0 = 4200
        w1 = 4600
    elif line == 'HeI5876':
        rest_wavelength = 5876.
        w0 = 5000
        w1 = 6000
    elif line == 'HeI6678':
        rest_wavelength = 6678.
        w0 = 6100 #6400
        w1 = 6450  # 6450 ; for -10days: 6300
    elif line == 'HeI7065':
        rest_wavelength = 7065
        w0 = 6700
        w1 = 7050
    elif line == 'HeI7281':
        rest_wavelength = 7281
        w0 = 7000
        w1 = 7300
    elif line == 'FeII5169':
        rest_wavelength = 5169
        w0 = 4600
        w1 = 5100  # 4950 ;for -10days: 4800  ;usual end: 5100
    elif line == 'FeII5018':
        rest_wavelength = 5018
        w0 = 4750
        w1 = 5000  # 4950
    elif line == 'Halpha':
        rest_wavelength = 6564
        w0 = 6000
        w1 = 6400  # 6400
        # w0 = 6200 & w1 = 6400
    elif line == 'IR_CaII':
        rest_wavelength = 8579.
        w0 = 7500
        w1 = 8500
    return w0, w1, rest_wavelength
```

```python
'''
Function: User can select two points on the plot of the SN spectrum. The points 
are supposed to mar blue and red side of selected absorption feature.
Inputs: line refers to the absorption line of interest, for example, 'HeI5876';
        row is the row number in the data table of that SN.
        Data of a SN is saved as a pickle. 
Outputs: left bound, i.e. point selected on the blue side of the absorption;
         right bound, i.e, point selected on the red side of the absorption
'''
def get_bounds(line, row):
    # Setting figure parameters
    plt.figure(figsize=(15,5))
    plt.clf()
    plt.minorticks_on()
    plt.tick_params('both', length=12, width=2, which='major', direction = 'in', top = True, right = False)
    plt.tick_params('x', length=8, width=1, which='minor', direction = 'in', top = True, right = False)
    plt.ylabel(r"Flux [$ergs^{-1}cm^{-2}\AA^{-1}]$", fontsize = 18)
    plt.xlabel(r"Rest Wavelength [$\AA$]", size = 18)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    title = "%s  age = %.1f days" %(sn_name, float(df['age'][row]))
    plt.title(title, fontdict={'fontsize': 15})  
    
    wv = df['wv'][row]
    dat = df['f_ft'][row]
    wv_l, wv_h, lam_s = select_line(line)
    msk = np.logical_and(wv > wv_l, wv < wv_h)
    wv = wv[msk]
    dat = dat[msk] 

    plt.plot(wv, dat, label = 'filtered spectrum')
    
    x = plt.ginput(2) 
    return x
```

```python
'''
Function: measuring velocity of an absorption line with uncertainty
Inputs: line refers to the absorption line of interest, for example, 'HeI5876';
        row is the row number in the data table of that SN.
        Data of a SN is saved as a pickle;
        bounds are the left and right bound of the absorption feature
Outputs: (absorption velocity, error in velocity)
'''
def measure_vel(line, row, bounds):
    wv = df['wv'][row] # wavelength array
    dat = df['f'][row] # flux array: as it is in the .ascii file downloaded from SNEx
    unc = df['unc'][row]# uncertainty array
    wv_l, wv_h, lam_s = select_line(line)
    msk = np.logical_and(wv > wv_l, wv < wv_h)
    wv = wv[msk]
    dat = dat[msk] 
    unc = unc[msk] 
    c_kms = 299792.47 # speed of light in km/s

    # Setting figure parameters
    plt.figure(figsize=(15,5))
    plt.clf()
    plt.minorticks_on()
    plt.tick_params('both', length=12, width=2, which='major', direction = 'in', top = True, right = False)
    plt.tick_params('x', length=8, width=1, which='minor', direction = 'in', top = True, right = False)
    plt.ylabel(r"Flux [$ergs^{-1}cm^{-2}\AA^{-1}]$", fontsize = 18)
    plt.xlabel(r"Rest Wavelength [$\AA$]", size = 18)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    title = "%s  age = %.1f days" %(sn_name, float(df['age'][row]))
    plt.title(title, fontdict={'fontsize': 15})
    plt.plot(wv, dat)
    plt.legend()
    
    #bounds = get_bounds(line, row)
    xlb = bounds[0][0] #left bound
    xrb = bounds[1][0] #right bound
    plt.scatter(bounds[0][0], bounds[0][1], c = 'red')
    plt.scatter(bounds[1][0], bounds[1][1], c = 'red')
    
    msk = np.logical_and(xlb < wv, xrb > wv)
    vel_arr = []
    nmc = 3000 # number of realizations
    
    for i in range(nmc):
        dat_tmp = np.copy(dat)
        wvl_bins = np.argwhere(msk)
        for wbin in wvl_bins:
            noise = np.random.normal(loc=0, scale=unc[wbin])
            dat_tmp[wbin] = dat_tmp[wbin] + noise
        model = np.poly1d(np.polyfit(wv[msk], dat_tmp[msk], 2))
        plt.plot(wv[msk], model(wv[msk]), c='k')
        coeffs = model.coeffs
        lam_o = - coeffs[1]/(2*coeffs[0]) # minimum wavelength 
        beta=(1-(lam_o/lam_s)**2)/((lam_o/lam_s)**2+1) # doppler formula
        vel = beta*c_kms
        vel_arr.append(vel)
    
    return (np.mean(vel_arr), np.std(vel_arr))
```

```python
row = 0
line = 'HeI5876'
bounds = get_bounds(line, row)
```

```python
(vel, vel_error) = measure_vel('HeI5876', row, bounds)
print((vel, vel_error))
```
